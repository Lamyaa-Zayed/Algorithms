```python
import itertools

LENGTH = 6  # Now I can change this global variable to solve the problem 
            # for any number of guards / prisoners.

assert LENGTH % 2 == 0 # Sanity check - gotta be an even number though. 


def genStates():
    """
    A function to generate all the possible states.
    UPDATED this to return an iterator, and to use itertools to create it.
    Input: None
    Output: all the possible states
    """
    side = ("E","W")   #East & Weast
    return itertools.product(side, repeat=LENGTH)


def genGraph(S):
    """
    A function to generate a graph that contains all the legal states and their next possible states to move
    Input: All the possible states
    Output: A graph of all the legal states and their next possible states to move

    """

    G = []
    graph = {}
    for state in S:
        if isLegal(state) == True:
            G.append(state)

    for legal_state in G:
        print("checking", legal_state)
        result1 = nextStates(legal_state, G)
        print("Next states:", result1)
        input()
        graph.update({legal_state: result1[1:]})  # add possible states to each legal states, put it in graph
    return graph

def isLegal(S):
    """
    A function to check if a state is legal or not.
    UPDATED to work regardless of how many wives & husbands there are.
    Input: None
    Output: Return True if a state is legal, else return False.

    """
    assert len(S) % 2 == 0  # Make sure we accept only odd-numbered arrays.
    husbands = [person for i, person in enumerate(S) if i % 2 == 0]
    wives = [person for j, person in enumerate(S) if j % 2 == 1]
    for k, wife in enumerate(wives):
        husband  = husbands[k]
        if not wife == husband:  # if she's not on the same side as her hubbie:
            if wife in husbands:  # and at least one other husband is there...
                return False
    return True


def nextStates(startnode, allstates):
    """
    A function to return a set of states that a state can move to.
    Input: A state and all the possible states
    Output: a set of states that a state can move to

    """

    possible = [startnode]
    count1 = startnode.count("E")
    count2 = startnode.count("W")
    for i in range(1, len(allstates)):
        if allstates[i].count("E") - count1 <= 2 and allstates[i].count("E") - count1 >= -2 and allstates[i].count(
                "W") - count2 <= 2 and allstates[i].count("W") - count2 >= -2:
            count = 0
            for j in range(0,len(startnode)):
                if allstates[i][j] == startnode[j]:
                    count += 1
            if count >= 4:  # Unlike MCGW problem, single unit's journey is unnecessary
                possible.append(allstates[i])
    return possible

def genShortestPath(graph, start, end, path=[]):

    """
    A function to find the shortest path in a graph, from the start to the destination.
    Input: A graph of all the legal states and their states next possible states to move, the starting point and the destination
    Output: the shortest path of the graph

    """

    path = path + [start]
    if start == end:
        return path
    if not (start in graph):
        return None
    shortestPath = None
    for node in graph[start]:
        print("Evaluating", node)
        if node not in path:
            newpath = genShortestPath(graph, node, end, path)
            if newpath:
                if not shortestPath or len(newpath) < len(shortestPath):
                    shortestPath = newpath
    return shortestPath



def Solver():
    S = genStates()
    G = genGraph(S)
    s = ''.join(["E"] * LENGTH)   # source node
    d = ''.join(['W'] * LENGTH)   # destination node
    result = genShortestPath(G, s, d)
    print(result)

Solver()
```

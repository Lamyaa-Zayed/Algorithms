```python
import networkx as nx 

dict1={
    '0':['1','3','8'],
    '1':['0','7'],
    '2':['3','5','7'],
    '3':['0','2','4'],
    '4':['3','8'],
    '5':['2','6'],
    '6':['5'], 
    '7':['1','2'],
    '8':['0','4'],    
}


G = nx.Graph(dict1)

# depth first search tree
T_dfs = list(nx.dfs_tree(G, source='0'))
print("dfs tree: ",T_dfs)

# breadth first search tree
T_bfs = list(nx.bfs_tree(G, source='0'))
print("bfs tree: ",T_bfs)

# draw the graph
nx.draw(G, with_labels=True, font_weight='bold')

```

```python
import networkx as nx 

m=0
graph={
    '0':['1','2'],
    '1':['0','3','5','6'],
    '2':['0','5','6'],
    '3':['1','7'],
    '4':['6','7'],
    '5':['1','2','7'],
    '6':['1','2','4','7'], 
    '7':['3','4','5','6'],
        
}
w=[8,9,8,1,7,9,9,6,4,1,4,7,7,7,6,2,9,4,7,6,4,7,2,6]

g= nx.Graph(graph)

nx.draw(g, with_labels=True, font_weight='bold')


for k,v in graph.items():
    for f in v:
        g.add_edge(k,f, weight = w[m])
        m+=1

 # dijkstra
shortest_path = nx.dijkstra_path(g, source='0', target='4')
print(shortest_path)
```

```python
import matplotlib.pyplot as plt
graph = {
    0: [2,1],
    1: [0,6,5,3],
    2: [5,6,0],
    3: [1,7],
    4: [6,7],
    5: [2,7,1],
    6: [2,1,7,4],
    7: [4,6,5,3],
    }
G = nx.Graph(graph)

G.add_edge(0,2,weight=9)
G.add_edge(0,1,weight=8)
G.add_edge(1,3,weight=1)
G.add_edge(1,5,weight=7)
G.add_edge(1,6,weight=9)
G.add_edge(2,6,weight=4)
G.add_edge(2,5,weight=6)
G.add_edge(5,7,weight=2)
G.add_edge(6,7,weight=6)
G.add_edge(3,7,weight=4)
G.add_edge(4,7,weight=7)

pos = nx.spring_layout(G)
nx.draw(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels, font_weight='bold')
plt.axis("off")
plt.show()
```

```python
distance, path = nx.single_source_dijkstra(G, source=0, target=7)
print("Shortest path between 0 and 7 is {}.".format(path))
print("The weight of the path is {}.".format(distance))
```
